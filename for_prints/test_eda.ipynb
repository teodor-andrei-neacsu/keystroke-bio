```python
import pandas as pd
```

```python
test_df = pd.read_csv("key_test.csv")
test_df.head()
```

```python
# transform dataframe to list of tuples
test_list = list(test_df.itertuples(index=False, name=None))
test_list
```

```python

typing_features = []

for idx, (tstamp, event, key) in enumerate(test_list):

    if event == 257:
        continue

    # get the release event
    for idx_rel, (tstamp_rel, event_rel, key_rel) in enumerate(test_list[idx + 1:]):
        if event_rel == 257 and key_rel == key:

            hl = tstamp_rel - tstamp
            
            # next pressed key
            for idx_next, (tstamp_next, event_next, key_next) in enumerate(test_list[idx + 1:]):
                if event_next == 256 and key_next != key:

                    il = tstamp_next - tstamp_rel
                    pl = tstamp_next - tstamp

                    # next release key
                    for idx_next_rel, (tstamp_next_rel, event_next_rel, key_next_rel) in enumerate(test_list[idx + 1:]):
                        if event_next_rel == 257 and key_next_rel == key_next:
                            rl = tstamp_next_rel - tstamp_rel
                            print("key: {}, hold: {}, inter: {}, press: {}, release: {}".format(key, hl, il, pl, rl))
                            typing_features.append((key, hl, il, pl, rl))
                            break
                    break
            break
```

```python
# create dataframe
feat_df = pd.DataFrame(typing_features, columns=["Key", "Hold Latency", "Inter-key Latency", "Press Latency", "Release Latency"])
feat_df
```

```python
feat_df.to_csv("typing_features.csv", index=False)
```

```python
# transform delta time to seconds
feat_df["Hold Latency"] = feat_df["Hold Latency"] * 0.001
feat_df["Inter-key Latency"] = feat_df["Inter-key Latency"] * 0.001
feat_df["Press Latency"] = feat_df["Press Latency"] * 0.001
feat_df["Release Latency"] = feat_df["Release Latency"] * 0.001
```

```python
# replae "Space" with " "
feat_df["Key"] = feat_df["Key"].replace("Space", " ")

# transform key to ascii
feat_df["Key"] = feat_df["Key"].apply(lambda x: ord(x))
```

```python
feat_df
```

```python
feat_df.to_excel("typing_features.xlsx", index=False)
```

```python
# normalize data
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
feat_df[["Hold Latency", "Inter-key Latency", "Press Latency", "Release Latency"]] = scaler.fit_transform(feat_df[["Hold Latency", "Inter-key Latency", "Press Latency", "Release Latency"]])

# transform ascii to [0, 1] range
feat_df["Key"] = feat_df["Key"].apply(lambda x: x / 255)
```
